# Loading Dataset To Pandas Dataframe

```python
credit_card_data = pd.read_csv("creditcard.csv")
```

## Distribution of Legit and Fraudulent Transaction

```python
credit_card_data['Class'].value_counts()
```

### This dataset is highly Unbalanced
### 0 --> Normal Transaction
### 1 --> Fraudulent Transaction

```python
# Seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

## Statistical Measures for the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values of both the transaction
credit_card_data.groupby('Class').mean()
```

## Under Sampling

### Building a dataset containing similar distribution of legit and fraud transactions

```python
# Number of Fraud transaction = 492
legit_sample = legit.sample(n=492)
```

### Concatenating the two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## Splitting the Data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Splitting the Data into Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

# Model Training

## Logistic Regression Model

```python
model = LogisticRegression(max_iter=5000, solver='lbfgs')
```

```python
# Training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

### Model Evaluation

```python
#Accuracy Score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training Data: ", training_data_accuracy)
```

```python
#Accuracy score on Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test Data: ", test_data_accuracy)
```
